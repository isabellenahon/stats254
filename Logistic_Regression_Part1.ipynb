```r
#recode the status variable into 1's and 0's

mydata$cancer_status <- recode(mydata$status, 'M' = 1, 'B' = 0)

table(mydata$status)
table(mydata$cancer_status)
```

```r
mydata %>%
ggplot() +
  aes( x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5)

```

```r
# Fit a logistic regression model
model <- glm(cancer_status ~ radius, data = mydata, family = binomial)

# Add the logistic regression curve to the plot
mydata %>%
  ggplot() +
  aes(x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5) +
  stat_smooth(method = "glm",
              method.args = list(family = "binomial"),
              se = FALSE, color = "red")


# Find the equation of the fitted curve
summary(model)

# Extract coefficients
intercept <- coef(model)[1]
slope <- coef(model)[2]

# Create a function for the logistic curve
logistic_curve <- function(x) {
  1 / (1 + exp(-(intercept + slope * x)))
}
```

fitted curve:

$ \hat p = \frac{e^{-15.2 + 1.03radius}}{1 + e^{-15.2 + 1.03radius}}$

where $\hat p$ is the predicited probablity from this model.

How should we Interprete 'slope'?

- First acknoaldge the sign of the 'slope' paramter.
  - In this case, the positive sign of the 'slope' indicates that probability increases as we move from small  to larger radiuses.
- In order to think about rate of change in probabolty, we need to make some judgement on the x values we use.

```r
mynewdata <- data.frame(radius = c(14,15,20,21))
prediction <- predict(model, newdata = mynewdata, type = 'response')
prediction
```
